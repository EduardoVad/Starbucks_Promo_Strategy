```python
training_df['Promotion'] = training_df['Promotion'].map({'Yes': 1, 'No': 0})
print(training_df['Promotion'].dtype)
```

```python
treat = training_df[training_df['Promotion'] == 1]
ctrl = training_df[training_df['Promotion'] == 0]
print(len(treat))
print(len(ctrl))
```

```python
purch_treat = len(treat[treat['purchase'] == 1])
cust_treat  = len(treat)
purch_ctrl  = len(ctrl[ctrl['purchase'] == 1])
cust_ctrl   = len(ctrl)

print(purch_treat, cust_treat, purch_ctrl, cust_ctrl)
```

```python
revenue     = 10
promo_cost  = 0.15


IRR = (purch_treat / cust_treat) - (purch_ctrl / cust_ctrl)
print('IRR:')
print(IRR)

NIR = ((revenue * purch_treat) - (promo_cost * cust_treat)) - (revenue * purch_ctrl)
print('NIR:')
print(NIR)
```

```python
def promotion_strategy(df):
    promote = (
        ((df['V1'] == 0) | (df['V1'] ==2)) &
        (df['V4'] == 2) & 
        ((df['V5'] == 3) | (df['V5'] ==1))
    )
    return promote.astype(int)
```

```python
result = promotion_strategy(training_df)
print(f"Customers targeted: {result.sum()}")
print(f"Percentage of total: {result.sum() / len(training_df) * 100:.2f}%")
```

```python
targeted = training_df[result == 1]

treat_t = targeted[targeted['Promotion'] == 1]
ctrl_t  = targeted[targeted['Promotion'] == 0]
```

```python
purch_treat_t = len(treat_t[treat_t['purchase'] == 1])
cust_treat_t  = len(treat_t)
purch_ctrl_t  = len(ctrl_t[ctrl_t['purchase'] == 1])
cust_ctrl_t   = len(ctrl_t)

IRR_targeted = (purch_treat_t / cust_treat_t) - (purch_ctrl_t / cust_ctrl_t)
NIR_targeted = (revenue * purch_treat_t - promo_cost * cust_treat_t) - (revenue * purch_ctrl_t)

print(f"Targeted IRR: {IRR_targeted:.4f}")
print(f"Targeted NIR: {NIR_targeted:.2f}")
```
